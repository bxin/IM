```python
%matplotlib inline
import os
from aosM2 import aosM2
import matplotlib.pyplot as plt
```

```python
debugLevel = 2
M2 = aosM2(debugLevel)
```

```python
zAngle = 0.81858933  # radian, = 46.9 degree
M2.printthz = M2.getPrintthz(zAngle)
```

```python
plt.scatter(M2.bx, M2.by, s=25, c=M2.printthz*1e6, marker='.',
                             edgecolor='none')
plt.axis('equal')
outerR = 1.71
plt.xlim(-outerR, outerR)
plt.ylim(-outerR, outerR)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.colorbar()
plt.title('M2 Gravitational Print Through (micron)') #micron because we have *1e6 with plt.scatter
```
