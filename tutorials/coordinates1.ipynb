```python
%matplotlib inline
import numpy as np
import aosCoTransform as ct
import matplotlib.pyplot as plt
```

```python
aa = np.loadtxt('../data/M1M3/M1M3_1um_156_grid.DAT')
```

```python
#aa has 5256 rows, for 5256 FEA nodes on M1M3, and 159 columns
#1st column is node ID, 2nd column is x-coordinate for each node, 3rd column is y-coordinate, 
# 4-159 are z-coordinate for each of the 156 bending modes
# these coordinates are all in the M1M3 coordinate system, with z pointing to the sky.
aa.shape
```

```python
bx = aa[:, 1]
by = aa[:, 2]
bz = aa[:, 3:]
```

```python
#bx1, by1, bz1 are the same coordinates in the ZEMAX cooridnate system (z pointing into M1M3)
bx1, by1, bz1 = ct.M1CRS2ZCRS(bx,by,bz)
```

```python
plt.scatter(bx1, by1, s=25, c=bz1[:,2]*1e6, marker='.',
                             edgecolor='none')
plt.axis('equal')
outerR = 4.18
plt.xlim(-outerR, outerR)
plt.ylim(-outerR, outerR)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.colorbar()
plt.title('M1M3 bending mode #3 (micron)') #micron because we have *1e6 with plt.scatter
```

```python
#same thing for M2 bending modes below
aa = np.loadtxt('../data/M2/M2_1um_grid.DAT')
```

```python
#aa has 9084 rows, for 9084 FEA nodes on M1M3, and 22 columns
#1st column is x-coordinate for each node, 2nd column is y-coordinate, 
# 3-22 are z-coordinate for the first 20 low order bending modes
# these coordinates are all in the M2 coordinate system, with z pointing to the sky.
aa.shape
```

```python
bx = aa[:, 0]
by = aa[:, 1]
bz = aa[:, 2:]
```

```python
#bx2, by2, bz2 are the same coordinates in the ZEMAX cooridnate system (z pointing into M1M3)
bx2, by2, bz2 = ct.M2CRS2ZCRS(bx,by,bz)
```

```python
plt.scatter(bx2, by2, s=25, c=bz2[:,3]*1e6, marker='.',
                             edgecolor='none')
plt.axis('equal')
outerR = 1.71
plt.xlim(-outerR, outerR)
plt.ylim(-outerR, outerR)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.colorbar()
plt.title('M2 bending mode #4 (micron)') #micron because we have *1e6 with plt.scatter
```
